```python
import numpy as np
import matplotlib.pyplot as plt
```

Letra A

```python
## Dados e Condições iniciais
m = 2.0
x0 = 1.0
v0 = 2.0
b = 0.1
ke = 5.0


def f(t, x, v):
    return v 

def g(t, x, v): # (independente, dependente, derivada)
    func2 = -(ke/m)*x - (b/m)*v  #v = dx/dt
    return func2


# Malha de Pontos
t = np.linspace(0.0, 12.0, 12000) # tempo
h = (t[-1]-t[0])/len(t) # Passo (precisão)


# Obter pontos
xm = [x0]
vm = [v0]
# Método de Euler Melhorado
for k in range(0, len(t)-1):
    k1 = f(t[k], xm[k], vm[k])
    k2 = f(t[k]+h, xm[k]+(h*k1), vm[k]+(h*k1))
    l1 = g(t[k], xm[k], vm[k])
    l2 = g(t[k]+h, xm[k]+(h*l1), vm[k]+(h*l1))
    vm.append(vm[k] + (h/2)*(l1 + l2))
    xm.append(xm[k] + (h/2)*(k1 + k2))


# Posição em determinado tempo:
def posicao(tm, x, h): # tempo, posições, passo
    if (tm == 0):
        posx = 0
    else:
        posx = x[int(tm/h)-1]
    
    posicao = posx
    return posicao


# Exemplo:
tempo = 5.0
print(posicao(tempo, xm, h)) 
```

Letra B

```python
## Calculando sem resistência

def fr(t, x, v):
    return v 

def gr(t, x, v): # (independente, dependente, derivada)
    func2 = -(ke/m)*x
    return func2


# Obter pontos
x = [x0]
v = [v0]
# Método de Euler Melhorado
for k in range(0, len(t)-1):
    k1 = fr(t[k], x[k], v[k])
    k2 = fr(t[k]+h, x[k]+(h*k1), v[k]+(h*k1))
    l1 = gr(t[k], x[k], v[k])
    l2 = gr(t[k]+h, x[k]+(h*l1), v[k]+(h*l1))
    v.append(v[k] + (h/2)*(l1 + l2))
    x.append(x[k] + (h/2)*(k1 + k2))
```

```python

plt.plot(t, xm, color='green', label='Com amortecimento')
plt.plot(t, x, color='red', label='Sem amortecimento')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid()
plt.show()
```
