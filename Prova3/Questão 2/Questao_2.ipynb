```python
import numpy as np
import matplotlib.pyplot as plt
```

Letra A

```python
# Definindo f
def f(x, y):
    return y - x**2 +1 

# Malha de Pontos
qtd = 15 #**
x = np.linspace(0.0, 2.0, qtd)
h = (x[-1]-x[0])/len(x) # Passo (precisão)

y0 = 0.5 # Valor inicial


### Solução Numérica por Euler Melhorado
y1 = [y0]
for k in range(0, len(x)-1):
    k1 = f(x[k], y1[k])
    k2 = f(x[k]+h, y1[k]+h*k1)
    
    y1.append(y1[k] + (h/2)*(k1+k2))


### Solução Numérica por Runge-Kutta de ordem 4 (método 2)
y2 = [y0]
for k in range(0, len(x)-1):
    k1 = f(x[k], y2[k])
    k2 = f(x[k]+(1/3)*h, y2[k]+(1/3)*h*k1)
    k3 = f(x[k]+(2/3)*h, y2[k]-(1/3)*h*k1+h*k2)
    k4 = f(x[k]+h, y2[k]+h*k1-h*k2+h*k3)

    y2.append(y2[k]+(h/8)*(k1+3*(k2+k3)+k4))


plt.plot(x, y1, color='red', label='Euler Melhorado')
plt.plot(x, y2, color='blue', label='Runge-Kutta 4')
plt.legend()
plt.grid()
plt.show()


# Encontrando um determinado ponto 
def posicao(tm, x, h): # ponto, posições, passo
    if (tm == 0):
        posx = 0
    else:
        posx = x[int(tm/h)-1]
    
    posicao = posx
    return posicao

xm = 2.0
print(f'Método Euler Melhorado: x = {xm}; y = {posicao(xm, y1, h)}')
print(f'Método Runge-Kutta de ordem 4: x = {xm}; y = {posicao(xm, y2, h)}')



#** Escolhi uma baixa quantidade de pontos para realçar as distorções
```

Letra B

```python
## Solução exata:
def sol(x):
    part1 = (x+1)**2
    part2 = (0.5)*np.exp(x)
    return part1 - part2

y3 = []
for i in range(0, len(x)):
    y3.append(sol(x[i]))


## Gráficos

plt.plot(x, y1, color='red', label='Euler Melhorado')
plt.plot(x, y2, color='blue', label='Runge-Kutta 4')
plt.plot(x, y3, color='green', label='Solução Analítica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()


## Comparando resultados:
xm = 2.0
print(f'Método Euler Melhorado: x = {xm}; y = {posicao(xm, y1, h)}')
print(f'Método Runge-Kutta de ordem 4: x = {xm}; y = {posicao(xm, y2, h)}')
print(f'Solução Analítica: x = {xm}; y = {sol(xm)}')
```

Há uma distorção muito maior de um método numérico para a solução analítica do que de um método numérico para outro método numérico. Essa distorção aumenta quanto mais o cálculo se distancia do valor inicial, mas pode ser reduzida aumentando a quantidade de pontos no intervalo x, consequentemente reduzindo o tamanho de cada passo h.
