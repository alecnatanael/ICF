```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing(order='grlex')

#Símbolos
g, v0, k, t, m = sp.symbols('g, v0, k, t, m')

#Função
y = sp.Function('y')
v = sp.Function('v')
```

```python
#Equação
eqy = sp.Eq(y(t).diff(t,2), +g -k*y(t).diff(t,1))
eqy
eqv = eqy.subs({y(t).diff(t,1): v(t)})
eqv
```

```python
#Resolução com condições iniciais:
sol = sp.dsolve(eqv, ics={v(0):v0})
sol
```

```python
#Lambdificação e Gráficos

tl = np.linspace(0,500,1000)

y_t = sp.lambdify(t, sol.rhs.subs({g:9.8, v0:0, k:0.08}))
Yp = y_t(tl)

x_t = tl
Xp = x_t

#plot

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(Xp, Yp, color='red')
plt.grid()
plt.show()
```

```python
#Qual o valor da velocidade terminal?

terminal = 1
for v in Yp:
    if v == terminal:
        terminal = v
        break
    terminal = v

print(f'{terminal:.2f} m/s')
```
