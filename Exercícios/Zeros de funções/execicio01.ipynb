```python
import numpy as np
import matplotlib.pyplot as plt
import sys

#Funções
'''
def f(x):
    # x - cosx = 0
    return x - np.cos(x)
'''
    
def f(x):
    # 2cos(x)-(e^x)/2 = 0
    return 2*np.cos(x) - pow(np.e, x)/2
```

```python
# Análise Gráfica:
start = int(input('Qual o x de início? '))
end = int(input('Qual o x final?' ))

X = list(range(start, end+1))
Y = [f(i) for i in X]

plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
#Condições iniciais
erro = 0.000001
a = -2
b = 0
i = 0

#Teorema de Bolzano
while (b - a) > erro:
    x = (b + a)/2
    if f(x)*f(a) > 0:
        a = x
    elif f(x)*f(b) > 0:
        b = x
    else:
        print(x)
        exit()
    i += 1
    print(f'Iteração {i} ; x = {x} ; f(x) = {f(x)} ; b-a = {b-a}')
```

```python
#Condições iniciais
erro = 0.000001
a = -2
b = 0
i = 0

# Método da Bissecção
while (b-a) > erro:
    x = (b+a)/2
    if (f(a) < 0 and f(b) > 0 and f(x) > 0) or (f(a) > 0 and f(b) < 0 and f(x) < 0):
        b = x
    elif (f(a) < 0 and f(b) > 0 and f(x) < 0) or (f(a) > 0 and f(b) < 0 and f(x) > 0):
        a = x
    else:
        print(f'O valor de x é {x}')
        sys.exit()

    i += 1
    print(f'Iteração {i} ; x = {x} ; f(x) = {f(x)} ; b-a = {b-a}')
```
