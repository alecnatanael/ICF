```python
import numpy as np
import matplotlib.pyplot as plt
import sys as sys

### Funções

def f(x):

    return x*np.log(x) -1


def df(x):

    return 1 + np.log(x)

    
def g(x):

    return np.tanh(1.5*x) -x


def dg(x):

    return (3/(2*pow(np.cosh(1.5*x),2))) - 1


def phif(x):

    return x - f(x)/df(x)


def phig(x):

    return x - g(x)/dg(x)


def phisf(x, x0):
    df = (f(x) - f(x0))/(x-x0)
    return x - f(x)/df


def phisg(x, x0):
    dg = (g(x) - g(x0))/(x-x0)
    return x - g(x)/dg
```

```python
# Gráfico para f(x):
start = int(input('Qual o x de início? '))
end = int(input('Qual o x final?' ))
pres = int(input('Qual a precisão? '))

X = list()

max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    X.append(x)
    x += 1/pres

Y = [f(i) for i in X]

plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
### Método da Bissecção para f(x):

# Condições iniciais
erro = 0.000001
a = 1.5
b = 2
i = 0

# Execução
while (b-a) > erro:
    x = (b+a)/2
    if (f(a) < 0 and f(b) > 0 and f(x) > 0) or (f(a) > 0 and f(b) < 0 and f(x) < 0):
        b = x
    elif (f(a) < 0 and f(b) > 0 and f(x) < 0) or (f(a) > 0 and f(b) < 0 and f(x) > 0):
        a = x
    else:
        print(f'Iteração {i} ; x = {x} ; g(x) = {g(x)} ; b-a = {b-a}')
        sys.exit()

    i += 1
    print(f'Iteração {i} ; x = {x} ; f(x) = {f(x)} ; b-a = {b-a}')
```

```python
### Método Newton-Raphson para f(x):

#Condições iniciais
erro1 = erro2 = 0.0000001
x0 = 1.5
k = 0

if abs(f(x0)) < erro1:
    k += 1
    print(f'Iteração {k} ; x = {x0} ; f(x) = {f(x0)}')
    sys.exit()

k += 1
while True:
    x = phif(x0)
    print(f'Iteração {k} ; x = {x} ; f(x) = {f(x)}')
    
    if abs(f(x)) < erro1 or abs(x-x0) < erro2:
        sys.exit()

    x0 = x
    k += 1
```

```python
### Método das Secantes para f(x):

#Condições iniciais
erro1 = erro2 = 0.0000001
x0 = 1.5    #Precisa ser menor do que a raiz esperada!
x = 1.8     #Precisa ser maior do que a raiz esperada!
k = 0

if abs(f(x0)) < erro1:
    k += 1
    print(f'Iteração {k} ; x = {x0} ; f(x) = {f(x0)}')
    sys.exit()

k += 1
while True:
    x = phisf(x, x0)
    print(f'Iteração {k} ; x = {x} ; f(x) = {f(x)}')

    if abs(f(x)) < erro1 or abs(x-x0) < erro2:
        break

    m = (x + x0)/2

    if (f(x0) < 0 and f(x) > 0 and f(m) > 0) or (f(x0) > 0 and f(x) < 0 and f(m) < 0):
        x = m
    elif (f(x0) < 0 and f(x) > 0 and f(m) < 0) or (f(x0) > 0 and f(x) < 0 and f(m) > 0):
        x0 = m

    k += 1
```
