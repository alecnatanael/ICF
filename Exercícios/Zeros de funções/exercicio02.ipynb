```python
import numpy as np
import matplotlib.pyplot as plt
import sys as sys

def f(x):
    # x^3 - 9x + 3 = 0
    return pow(x,3) - 9*x + 3

def phi(x):
    # ((x^3)/9)+(1/3)
    return (pow(x,3)/9)+(1/3)
```

```python
# Análise Gráfica:
start = int(input('Qual o x de início? '))
end = int(input('Qual o x final?' ))

X = list(range(start, end+1))
Y = [f(i) for i in X]

plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
#Condições iniciais
erro = 0.000001
a = 0
b = 1
i = 0

#Teorema de Bolzano
while (b - a) > erro:
    x = (b + a)/2
    if f(x)*f(a) > 0:
        a = x
    elif f(x)*f(b) > 0:
        b = x
    else:
        print(x)
        exit()
    i += 1
    print(f'Iteração {i} ; x = {x} ; f(x) = {f(x)} ; b-a = {b-a}')
```

```python
#Condições iniciais
erro = 0.000001
a = 0
b = 1
i = 0

# Método da Bissecção
while (b-a) > erro:
    x = (b+a)/2
    if (f(a) < 0 and f(b) > 0 and f(x) > 0) or (f(a) > 0 and f(b) < 0 and f(x) < 0):
        b = x
    elif (f(a) < 0 and f(b) > 0 and f(x) < 0) or (f(a) > 0 and f(b) < 0 and f(x) > 0):
        a = x
    else:
        print(f'O valor de x é {x}')
        sys.exit()

    i += 1
    print(f'Iteração {i} ; x = {x} ; f(x) = {f(x)} ; b-a = {b-a}')
```

```python
#Condições iniciais
erro1 = erro2 = 0.000001
x0 = 1
k = 0

# Método iterativo linear (MIL)
if abs(f(x0)) < erro1:
    k += 1
    print(f'Iteração {k} ; x = {x0} ; f(x) = {f(x0)}')
    sys.exit()

k += 1
while True:
    x = phi(x0)
    print(f'Iteração {k} ; x = {x} ; f(x) = {f(x)}')
    
    if abs(f(x)) < erro1 or abs(x-x0) < erro2:
        sys.exit()

    x0 = x
    k += 1
    
```
