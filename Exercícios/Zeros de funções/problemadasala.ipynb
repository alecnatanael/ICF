```python
import numpy as np
import matplotlib.pyplot as plt
import sys as sys

theta = np.radians(80)
#theta = 1.39
v2gr = 1.25
A = np.tan(theta/2)
R = 6371000
g = 9.81

#Funções

def f(alpha): #Função baseada na equação dada em sala, o zero dessa função revela o valor de alpha necessário
    falpha = A*((1/v2gr)-pow(np.cos(alpha),2)) - (np.sin(alpha)*np.cos(alpha))
    return falpha

def bissec(a, b, erro, M):

    while (b-a) > erro:
        E = (b+a)/2
        if (f(a) < M and f(b) > M and f(E) > M) or (f(a) > M and f(b) < M and f(E) < M):
            b = E
        elif (f(a) < M and f(b) > M and f(E) < M) or (f(a) > M and f(b) < M and f(E) > M):
            a = E
        else:
            break
    
    return E
```

```python
# Gráfico para f(alpha):
start = int(input('Qual o alpha de início? '))
end = int(input('Qual o alpha final?' ))
pres = int(input('Qual a precisão? '))

X = list()

max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    X.append(x)
    x += 1/pres

Y = [f(np.radians(i)) for i in X]


plt.xlabel("alpha (graus)", fontsize=14)
plt.ylabel("f(alpha)", fontsize=14)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
### Aproximando o valor de alpha

#Condições
erro = 0.000001
a = np.radians(58)
b = np.radians(60)
obj = 0

alpha = bissec(a, b, erro, obj)
print(np.degrees(alpha))
```
