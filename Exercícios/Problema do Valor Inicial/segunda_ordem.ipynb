```python
import numpy as np
import matplotlib.pyplot as plt
```

Euler Simples (2Â° Ordem)

```python
## Problemas do Slide:


### 1
def f(x, y, z):
    return z

def g(x, y, z): # (independente, dependente, derivada)
    func2 = 4*z - 3*y - x
    return func2


# Malha de Pontos e condiÃ§Ãµes iniciais
x = np.linspace(0.0, 1.0, 100)

y0 = 4/9
z0 = 7/3

h = (x[-1]-x[0])/len(x) # Passo (precisÃ£o)

y = [y0]
z = [z0]
for k in range(0, len(x)-1):
    z.append(z[k] + h*(g(x[k], y[k], z[k])))
    y.append(y[k] + h*f(x[k], y[k], z[k]))
    


# grÃ¡fico
plt.plot(x, y, color='blue', label='Euler Simples')
plt.legend()
plt.grid()
plt.show()
```

Euler Melhorado (2Â° ordem)

```python
### 2
def f(x, y, z): # y' (independente, dependente, derivada)
    return z

def g(x, y, z): # z' (independente, dependente, derivada)
    func2 = y + np.exp(x)
    return func2


# Malha de Pontos e condiÃ§Ãµes iniciais
x = np.linspace(0.0, 1.0, 100)

y0 = 1
z0 = 0

h = (x[-1]-x[0])/len(x) # Passo (precisÃ£o)

y = [y0]
z = [z0]
# MÃ©todo de Euler Melhorado
for k in range(0, len(x)-1):
    k1 = f(x[k], y[k], z[k])
    k2 = f(x[k]+h, y[k]+h*k1, z[k] + h*k1)
    l1 = g(x[k], y[k], z[k])
    l2 = g(x[k]+h, y[k]+h*l1, z[k] + h*l1)
    z.append(z[k] + (h/2)*(l1 + l2))
    y.append(y[k] + (h/2)*(k1 + k2))
    


# grÃ¡fico
plt.plot(x, y, color='blue', label='Euler Simples')
plt.legend()
plt.grid()
plt.show()
```
