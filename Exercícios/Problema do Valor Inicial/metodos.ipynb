```python
import numpy as np
```

Euler Simples

```python
y0 = 1 # Valor inicial
y = [y0] 


# Malha de Pontos
x = np.linspace(0.0, 2.0, 50)
h = (x[-1]-x[0])/len(x) # Passo (precisão)

# Definindo f
def f(x, y):
    return x - 2*y

# Aplicando euler-simples
for k in range(0, len(x)-1):
    y.append(y[k] + h*f(x[k], y[k]))


#Solução analítica:
def sol(x):
    return (5*np.exp(-2*x) + 2*x - 1)/4


import matplotlib.pyplot as plt

xm = np.linspace(x[0], x[-1], len(x))
ym = []

for i in range(0, len(xm)):
    ym.append(sol(xm[i]))


plt.plot(x, y, color='blue', label='Euler Simples')
plt.plot(xm, ym, color='red', label='Solução Analítica')
plt.legend()
plt.grid()
plt.show()
```

Runge-Kutta Ordem 2 (Euler Modificado)

```python
y0 = 3 # Valor inicial

h = 0.125 # Passo (precisão)

x0 = 0.0
xf = 5.0
N = (xf - x0)/h

x = np.linspace(x0, xf, int(N)) 

 


# Definindo f
def f(x, y):
    part1 = (2*y)/(x+1)
    part2 = (x + 1)**3
    return part1+part2

y = [y0] # Valor inicial
for k in range(0, len(x)-1):
    k1 = f(x[k], y[k])
    k2 = f(x[k]+(1/2)*h, y[k]+(1/2)*h*k1)
    
    y.append(y[k] + h*k2)


import matplotlib.pyplot as plt

plt.plot(x, y, color='blue', label='Euler Modificado')
plt.legend()
plt.grid()
plt.show()
```

Runge-Kutta Ordem 3 (Heun e Nystrom)

```python
### Heun ###

y0 = 3.0 # Valor Inicial

h = 0.125 # Passo (precisão)

x0 = 0.0
xf = 5.0
N = (xf - x0)/h

x1 = np.linspace(x0, xf, int(N)) 

# Definindo f
def f(x, y):
    part1 = (2*y)/(x+1)
    part2 = (x + 1)**3
    return part1+part2


y1 = []
y1.append(y0)

for k in range(0, len(x1)-1):
    k1 = f(x1[k], y1[k])
    k2 = f(x1[k]+(h/3), y1[k]+(h/3)*k1)
    k3 = f(x1[k]+(2/3)*h, y1[k]+(2/3)*h*k2)
    
    y1.append(y1[k] + (h/4)*(k1+3*k3))

```

```python
### Nystrom ###

y0 = 3.0 # Valor Inicial

h = 0.125 # Passo (precisão)

x0 = 0.0
xf = 5.0
N = (xf - x0)/h

x2 = np.linspace(x0, xf, int(N))


# Definindo f
def f(x, y):
    part1 = (2*y)/(x+1)
    part2 = (x + 1)**3
    return part1+part2


y2 = []
y2.append(y0)

for k in range(0, len(x2)-1):
    k1 = f(x2[k], y2[k])
    k2 = f(x2[k]+(2/3)*h, y2[k]+(2/3)*h*k1)
    k3 = f(x2[k]+(2/3)*h, y2[k]+(2/3)*h*k2)
    
    y2.append(y2[k] + (h/4)*(k1+(3/2)*(k2+k3)))

```

```python
plt.plot(x2, y2, color='blue', label='Nystrom')
plt.plot(x1, y1, color='Red', label='Heun')
plt.legend()
plt.grid()
plt.show()
```

Runge-Kutta de ordem 4

```python
y0 = 2
h = 0.1

x0 = 0.0
xf = 5.0
N = (xf - x0)/h

x = np.linspace(x0, xf, int(N)) # Baseado em h

# Definindo f
def f(x, y):
    return (-y) + x + 2


# Segundo método:

y = []
y.append(y0)
for k in range(0, len(x)-1):
    k1 = f(x[k], y[k])
    k2 = f(x[k]+(1/3)*h, y[k]+(1/3)*h*k1)
    k3 = f(x[k]+(2/3)*h, y[k]-(1/3)*h*k1+h*k2)
    k4 = f(x[k]+h, y[k]+h*k1-h*k2+h*k3)

    y.append(y[k]+(h/8)*(k1+3*(k2+k3)+k4))


plt.plot(x, y, color='blue', label='Runge-Kutta 4')
plt.legend()
plt.grid()
plt.show()
```
