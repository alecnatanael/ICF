Exercício 1

```python
def interpolacao(x, fx): ## Obtém os coeficientes do polinômio Pn(x)
    grau = len(x)
    A = np.zeros((grau, grau))
    for i in range(0, grau):
        for j in range(0, grau):
            A[i][j] = x[i]**j

    B = np.copy(fx)

    coef = solucao_direta(A, B)


    return coef


x = [-1, 0, 3]
fx = [15, 8, -1]

a = interpolacao(x, fx)

print(a)
```

```python
def Polinomio(x, a): ## Retorna o valor correspondente a Pn(x) para um polinômio com os coeficientes do vetor "a"
    n = len(a)


    Pnx = 0
    for j in range(0, n):
        Pnx += a[j]*(x**j)
        
    
    return Pnx
```

```python
import matplotlib.pyplot as plt

xm = np.linspace(x[0], x[-1], 100) ## cria uma "malha fina" de 100 pontos do primeiro x ao último

fi = []

for i in range(0, len(xm)):
    fi.append(Polinomio(xm[i], a)) 


plt.scatter(x, fx, color='red', label='Dados exp.')
plt.plot(xm, fi, color='blue', label='Ajuste')
plt.grid()
plt.show()
```
