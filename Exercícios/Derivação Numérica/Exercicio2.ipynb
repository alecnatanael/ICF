```python
import numpy as np
import matplotlib.pyplot as plt

massa = 10
h = 0.001

def fy(t):
    return 30*t*np.sin(np.radians(60)) - (9.8*(t**2)/2)

def fx(t):
    return 30*t*np.cos(np.radians(60))

def d2x(t):
    fll = (1/(h**2))*(fx(t+h)-(2*fx(t))+fx(t-h))
    return fll

def d2y(t):
    fll = (1/(h**2))*(fy(t+h)-(2*fy(t))+fy(t-h))
    return fll

```

```python
### Gráficos da posição:

start = 0
end = 6
pres = 100

tempo = list()

max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    tempo.append(x)
    x += 1/pres


X = [fx(i) for i in tempo]
Y = [fy(i) for i in tempo]

plt.xlabel("x(t)", fontsize=14)
plt.ylabel("y(t)", fontsize=14)

plt.plot(X,Y, color='blue')
plt.grid()
plt.show()
```

```python
# Gráficos das acelerações:

start = 0
end = 6
pres = 100

tempo = list()
max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    tempo.append(x)
    x += 1/pres

Y1 = [d2x(i) for i in tempo] # no eixo x
Y2 = [d2y(i) for i in tempo] # no eixo y

plt.xlabel("tempo (s)", fontsize=14)
plt.ylabel("Acelerações", fontsize=14)

plt.plot(tempo,Y1, color='black', label='Eixo x')
plt.plot(tempo,Y2, color='red', label='Eixo y')
plt.legend()
plt.grid()
plt.show()

print(Y1)
print(Y2)
```

```python
# Como vimos, temos uma aceleração constante no eixo y, que é justamente a aceleração da gravidade! Logo só haverá uma força sobre a partícula no eixo y

# Força sobre a partícula

start = 0
end = 6
pres = 100

tempo = list()
max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    tempo.append(x)
    x += 1/pres

YFy = [massa*d2y(i) for i in tempo]
YFx = [massa*d2x(i) for i in tempo]
#YF = [massa*9.8 for i in tempo]

print(YFy)
print(YFx)

plt.xlabel("tempo (s)", fontsize=14)
plt.ylabel("Força (N)", fontsize=14)

plt.plot(tempo,YFy, color='red', label='Eixo y')
plt.plot(tempo,YFx, color='black', label='Eixo x')
plt.legend()
plt.grid()
plt.show()
```
