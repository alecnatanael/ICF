```python
import numpy as np
import math
import matplotlib.pyplot as plt

h = 0.001


def f(x):
    return np.exp(-x**2)

def d1(x):
    fl = (1/(2*h))*(f(x + h) - f(x - h))
    return fl

def d2(x):
    fll = (1/(h**2))*(f(x+h)-(2*f(x))+f(x-h))
    return fll

def H0(x):
    ##return np.exp(x**2)*f(x)
    return 1

def H1(x):
    return (-1)*np.exp(x**2)*d1(x)

def H2(x):
    return np.exp(x**2)*d2(x)

def Hn(x):
    Hs1 = H1(x)
    Hs2 = H2(x)
    H = [H0(x), Hs1, Hs2]
    for n in range(3,11):    
        H.append(2*x*Hs2 - 2*(n-1)*Hs1)
        Hs1 = Hs2
        Hs2 = H[n]
    
    return H

def psi(x, n):
    Hs = Hn(x)
    psis = []
    for i in range(0,11):
        psis.append(np.sqrt(1/((2**i)*math.factorial(i)*np.sqrt(np.pi)))*(Hs[i])*np.exp((-x**2)/2))
    return psis[n]
```

```python
#Plotar grficos de psi(x, n):

start = -4
end = 4
pres = 100

X = list()

max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    X.append(x)
    x += 1/pres

Y0 = [psi(i, 0) for i in X]
Y1 = [psi(i, 1) for i in X]
Y2 = [psi(i, 2) for i in X]
Y3 = [psi(i, 3) for i in X]
Y4 = [psi(i, 4) for i in X]
Y5 = [psi(i, 5) for i in X]
Y6 = [psi(i, 6) for i in X]
Y7 = [psi(i, 7) for i in X]
Y8 = [psi(i, 8) for i in X]
Y9 = [psi(i, 9) for i in X]
Y10 = [psi(i, 10) for i in X]

plt.xlabel("x", fontsize=14)
plt.ylabel("psi(x)", fontsize=14)

plt.plot(X,Y0, color='black', label='n=0')
plt.plot(X,Y1, color='red', label='n=1')
plt.plot(X,Y2, color='orange', label='n=2')
plt.plot(X,Y3, color='yellow', label='n=3')
plt.plot(X,Y4, color='pink', label='n=4')
plt.plot(X,Y5, color='brown', label='n=5')
plt.plot(X,Y6, color='green', label='n=6')
plt.plot(X,Y7, color='cyan', label='n=7')
plt.plot(X,Y8, color='blue', label='n=8')
plt.plot(X,Y9, color='gray', label='n=9')
plt.plot(X,Y10, color='lightgray', label='n=10')
#plt.legend()
plt.grid()
plt.show()
```
