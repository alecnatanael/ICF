MÃ©todo Gauss-Jacobi

```python
A = [[10, 2, 1],
    [1, 5, 1],
    [2, 3, 10]]

B = [7, -8, 6]

chute = [7/10, -8/5, 6/10]

eps = 0.0001

def gauss_jacobi(A, B, chute, eps):
    it = 1
    i = 0
    n = len(A)

    
    x = chute.copy()
    xk = chute.copy()

    print('xk = ', xk)
    print('x = ', x)

    while True:

        for i in range(0, n):   
            soma = 0
            for j in range(0, n):
                if j != i:
                    soma += (A[i][j])*(x[j])
                    
            xk[i] = (1/(A[i][i]))*(B[i] - soma)


        print('Iteracao = ', it)
        print('xk = ', xk)
        print('x = ', x)
        it += 1

        ds = []
        for m in range(0, n):
            ds.append(abs(xk[m]-x[m]))
        
        if (max(ds) < eps):
            break

        x = xk.copy()
        

    return xk
    


print(gauss_jacobi(A, B, chute, eps))
```

```python
A = [[10.0, 2.0, 1.0],
    [1.0, 5.0, 1.0],
    [2.0, 3.0, 10.0]]

B = [7.0, -8.0, 6.0]

vetG = [7/10, -8/5, 6/10]

nitmax = 20

def gaussjacobi(Ac, bc, vetGuess, nit):
    it = 0
    n = len(Ac)

    x = vetGuess
    xk = vetGuess

    while (it < nit):
        it += 1

        for i in range(1, n+1, 1):
            soma = 0.0
            for j in range(1, n+1, 1):
                if ((i-1) != (j-1)):
                    soma = soma + (Ac[i-1][j-1]*x[j-1])
                
            xk[i-1] = (1.0/Ac[i-1][i-1]*(bc[i-1]-soma))

        x = xk
        print('Iteracao = ', it)
        print('x = ', xk)
    
    return x


print(gaussjacobi(A, B, vetG, nitmax))
```

MÃ©todo Gauss-Seidel

```python
A = [[-7, 3, 2],
    [1, 2, -1],
    [1, 1, -2]]

B = [-2, 2, 0]

chute = [1/2, 1/2, 1/2]

eps = 0.000001

def gauss_seidel(A, B, chute, eps):
    it = 1
    i = 0
    n = len(A)

    
    x = chute.copy()
    xk = chute.copy()

    print('xk = ', xk)
    print('x = ', x)

    while True:

        for i in range(0, n):   
            soma = 0
            
            for j in range(0, n):
                if i > j:
                    soma += (A[i][j])*(xk[j])
                elif i < j:
                    soma += (A[i][j]*x[j])
            xk[i] = (1/(A[i][i]))*(B[i] - soma)


        print('Iteracao = ', it)
        print('xk = ', xk)
        print('x = ', x)
        it += 1

        ds = []
        
        for m in range(0, n):
            ds.append(abs(xk[m]-x[m]))
        
        if (max(ds) < eps):
            break

        x = xk.copy()
        

    return xk
    


print(gauss_seidel(A, B, chute, eps))
```
