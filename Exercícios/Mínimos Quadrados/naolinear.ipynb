```python
import numpy as np
from ICF2 import *
```

```python
x = [-1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
fx = [0.157, 0.234, 0.350, 0.522, 0.778, 1.162, 1.733, 2.586, 3.858]
```

```python
def funcgs2(xc): ## para g1 = 1; g2 = x;

    m = len(xc)

    g1 = np.array([1.0]*m)
    g2 = np.array(xc)

    gm = [g1, g2]

    return gm
```

```python
gs = funcgs2(x)
alphas = minimos_quadrados(x, fx, gs)
print (alphas)
```

```python
Beta1 = np.log(alphas[0])
Beta2 = alphas[1]

print(f'{Beta1:.2f} {Beta2:.2f}')
```

```python
import matplotlib.pyplot as plt

xm = np.linspace(x[0], x[-1], 100) ## cria uma "malha fina" de 100 pontos do primeiro x ao último


fi = [] ## Obtendo todos os phis

for i in range(0, len(xm)):
    fi.append(np.e**(Beta1 + Beta2*xm[i])) 


plt.scatter(x, fx, color='red', label='Dados exp.')
plt.plot(xm, fi, color='blue', label='Ajuste')
plt.grid()
plt.show()
```
