```python
### Constantes
N = 31
C = 1
m = 1
k = 6
omega = 2

kmomega = k/(m*(omega**2))
Cmomega = C/(m*(omega**2))
```

```python
### Gerando as matrizes correspondentes ao sistema linear obtido na letra A

MatrizA = np.zeros((N, N))
for i in range(0, N):
    for j in range(0, N):
        if ((i == 0) and (j == 0)) or ((i == N-1) and (j == N-1)):
            MatrizA[i][j] = 1 - kmomega
        elif ((i == 0) and (j == 1)) or ((i == N-1) and (j == N-2)):
            MatrizA[i][j] = kmomega
        elif i==j:
            MatrizA[i][j] = (1 - 2*kmomega)
            MatrizA[i][j-1] = kmomega
            MatrizA[i][j+1] = kmomega



MatrizB = [-Cmomega]
for i in range(1, N):
    MatrizB.append(0)


## Imprimindo as Matrizes
for i in range(0, len(MatrizA)):
    for j in range(0, len(MatrizA)):
        print(f'{MatrizA[i][j]:.2f} ', end="")
    print('\n')

print('\n')
for i in range(0, len(MatrizB)):
    print(f'{MatrizB[i]:.2f} ', end="")
```

```python
### Solução de sistemas lineares de forma direta:

def solucao_direta(A, B):
    n = len(A)

    # Criando um sistema triangular com o método da Eliminação de Gauss
    for k in range(0, n-1):
        for i in range(k+1, n):
            m = (A[i][k])/(A[k][k])

            for j in range(k, n):
                A[i][j] = A[i][j] - m*A[k][j]
            B[i] = B[i] - m*B[k]


    # Resolvendo o sistema triangular pelo método da substituição retroativa
    x = []

    for i in range(0, n):
        x.append(0)

    x[n-1] = B[n-1]/A[n-1][n-1]

    c = 1
    
    for i in range(n-2, -1, -1):
        soma = 0
        c += 1
        for j in range(n-1, i, -1):
            soma += (A[i][j]*x[j])
        x[n-c] = (B[i] - soma)/(A[i][i])

    return x # retorna o vetor solução do sistema
```

```python
### Resolvendo o sistema linear para obter o vetor "Az" com as amplitudes de oscilação de cada bolinha

Az = solucao_direta(MatrizA, MatrizB)

c = 1
for i in Az:
    print(f'A{c}: {i}')
    c += 1
```
