```python
import numpy as np
import matplotlib.pyplot as plt
import sys as sys

def f(h): #Equação em 0
    return ((1.2**2)/2*9.81*(1.8**2)*(0.6**2)) + 0.6 - ((1.2**2)/2*9.81*(1.8**2)*(h**2)) - 0.0075 - h

### Bissecção:
def bissec(a, b, erro, zero):
    k = 1

    while (b-a) > erro:
        E = (b+a)/2
        if (f(a) < zero and f(b) > zero and f(E) > zero) or (f(a) > zero and f(b) < zero and f(E) < zero):
            b = E
            print(f'Iteração {k} ; x = {E} ; f(x) = {f(E)}')
            k += 1
        elif (f(a) < zero and f(b) > zero and f(E) < zero) or (f(a) > zero and f(b) < zero and f(E) > zero):
            a = E
            print(f'Iteração {k} ; x = {E} ; f(x) = {f(E)}')
            k += 1
        else:
            break
    
    return E
```

```python
#Plotar gráfico de f(x):
start = 0
end = 1
pres = 100

X = list()

max = abs(start-end)*pres
x = start
for i in range(1, max+1):
    X.append(x)
    x += 1/pres

Y = [f(i) for i in X]

plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)

plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
# Método da Bissecção
raiz = bissec(0.5, 0.7, 0.000001, 0)
print(raiz) #A altura do canal é a raiz da equação dada
```
